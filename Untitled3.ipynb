```python
arr = list(map(int, input("Введите список: ").split()))
K = int(input("K = "))
L = int(input("L = "))

# Нумерация с 1, поэтому K-1:L
segment = arr[K-1:L]
average = sum(segment) / len(segment)

print(f"Среднее арифметическое = {average:.6f}")
```

```python
arr = list(map(int, input("Введите список: ").split()))
K = int(input("K = "))
L = int(input("L = "))

excluded_sum = sum(arr) - sum(arr[K-1:L])
excluded_count = len(arr) - (L - K + 1)

average = excluded_sum / excluded_count

print(f"Среднее арифметическое = {average:.6f}")
```

```python
arr = list(map(int, input("Введите список: ").split()))

# Четные по возрастанию индексов
for i in range(len(arr)):
    if arr[i] % 2 == 0:
        print(arr[i], end=' ')

# Нечетные по убыванию индексов
for i in range(len(arr)-1, -1, -1):
    if arr[i] % 2 != 0:
        print(arr[i], end=' ')
print()
```

```python
arr = list(map(int, input("Введите список: ").split()))

max_index = arr.index(max(arr))
product = 1

for i in range(max_index + 1, len(arr)):
    product *= arr[i]

print(f"Произведение = {product}")
```

```python
arr = list(map(int, input("Введите список: ").split()))

min_index = arr.index(min(arr))
max_index = arr.index(max(arr))

start = min(min_index, max_index)
end = max(min_index, max_index)

total = sum(arr[start:end+1])

print(f"Сумма = {total}")
```

```python
arr = list(map(int, input("Введите список: ").split()))

min_index = arr.index(min(arr))
max_index = arr.index(max(arr))

start = min(min_index, max_index)
end = max(min_index, max_index)

for i in range(start + 1, end):
    arr[i] = 0

print("Результат:", arr)
```

```python
arr = list(map(int, input("Введите список: ").split()))
value = int(input("На сколько увеличить четные числа: "))

for i in range(len(arr)):
    if arr[i] % 2 == 0:
        arr[i] += value

print("Результат:", arr)
```
