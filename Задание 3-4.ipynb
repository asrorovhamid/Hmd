```python
price = float(input("Цена за 1 кг: "))

kg = 1.0
while kg <= 2.01:  # 2.01 для учета погрешности float
    cost = price * kg
    print(f"{kg:.1f} кг — {cost:.2f} руб.")
    kg += 0.2
```
