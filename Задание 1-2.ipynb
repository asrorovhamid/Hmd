```python
# Вводим длины ребер
a = float(input("Введите длину ребра a: "))
b = float(input("Введите длину ребра b: "))
c = float(input("Введите длину ребра c: "))

# Вычисляем объем
V = a * b * c

# Вычисляем площадь поверхности
S = 2 * (a*b + b*c + a*c)

# Выводим результаты
print(f"\nПри a = {a}, b = {b}, c = {c}")
print(f"Объем V = a ⋅ b ⋅ c = {V}")
print(f"Площадь поверхности S = 2 ⋅ (a⋅b + b⋅c + a⋅c) = {S}")
```

```python
import math

# Вычисляем для разных радиусов
print("Длина окружности и площадь круга для разных радиусов:")
print("R\tL\t\tS")
print("-" * 35)

for R in [1, 2, 3, 5, 10]:
    L = 2 * math.pi * R
    S = math.pi * R**2
    print(f"{R}\t{L:.4f}\t{S:.4f}")
```

```python
import math

# Вычисляем для разных катетов
print("Гипотенуза и периметр для разных треугольников:")
print("a\tb\tc\t\tP")
print("-" * 40)

triangles = [(3, 4), (5, 12), (6, 8), (7, 24)]
for a, b in triangles:
    c = math.sqrt(a**2 + b**2)
    P = a + b + c
    print(f"{a}\t{b}\t{c:.4f}\t{P:.4f}")
```

```python
import math

x1, y1 = map(float, input("Введите x1 y1: ").split())
x2, y2 = map(float, input("Введите x2 y2: ").split())
x3, y3 = map(float, input("Введите x3 y3: ").split())

def distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

a = distance(x1, y1, x2, y2)
b = distance(x2, y2, x3, y3)
c = distance(x3, y3, x1, y1)

perimeter = a + b + c
p = perimeter / 2
area = math.sqrt(p * (p - a) * (p - b) * (p - c))

print(f"Периметр: {perimeter:.4f}")
print(f"Площадь: {area:.4f}")
```

```python
X = float(input("Кг шоколадных конфет (X): "))
A = float(input("Стоимость A руб: "))
Y = float(input("Кг ирисок (Y): "))
B = float(input("Стоимость B руб: "))

price_choc = A / X
price_iris = B / Y
ratio = price_choc / price_iris

print(f"1 кг шоколадных конфет: {price_choc:.2f} руб")
print(f"1 кг ирисок: {price_iris:.2f} руб")
print(f"Шоколадные конфеты дороже в {ratio:.2f} раз")
```

```python
A1, B1, C1 = map(float, input("Введите A1 B1 C1: ").split())
A2, B2, C2 = map(float, input("Введите A2 B2 C2: ").split())

D = A1 * B2 - A2 * B1
Dx = C1 * B2 - C2 * B1
Dy = A1 * C2 - A2 * C1

x = Dx / D
y = Dy / D

print(f"x = {x:.4f}")
print(f"y = {y:.4f}")
```
